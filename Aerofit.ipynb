```python
# Defining the numerical attributes you want to check for outliers
numerical_attributes = ['Age', 'Usage', 'Income', 'Fitness', 'Miles']

# Creating a boxplot for each numerical attribute
for attribute in numerical_attributes:
    df.boxplot(column=attribute)
    plt.title(f'Boxplot for {attribute}')
    plt.show()
```

```python
# Calculating the difference between the mean and median for each attribute
outliers = []
for attribute in numerical_attributes:
    mean = df[attribute].mean()
    median = df[attribute].median()
    difference = np.abs(mean - median)
    outliers.append({'Attribute': attribute, 'Mean': mean, 'Median': median, 'Difference': difference})

# Displaying the attributes with significant differences between mean and median
outliers_df = pd.DataFrame(outliers)
print("Attributes with significant differences between Mean and Median:")
print(outliers_df[outliers_df['Difference'] > 0])
```


The following attributes have significant differences between their mean and median values, indicating potential skewness in their distributions:

Age: The mean age (28.79) is higher than the median age (26.0), with a difference of approximately 2.79 years. This suggests a right-skewed distribution, where some customers are older than the majority.

Usage: The mean usage (3.46 times per week) is higher than the median usage (3.0 times per week), with a difference of approximately 0.46 times. This suggests that some customers use the treadmill more frequently than the majority.

Income: The mean income ($53,719.58) is higher than the median income ($50,596.5), with a difference of approximately $3,123.08. This indicates a right-skewed distribution, with some customers having higher incomes than the majority.

Fitness: The mean fitness rating (3.31) is higher than the median fitness rating (3.0), with a difference of approximately 0.31. This suggests that some customers rate their fitness higher than the majority.

Miles: The mean expected miles (103.19) is higher than the median expected miles (94.0), with a difference of approximately 9.19 miles. This indicates that some customers expect to walk or run more miles each week than the majority.

These differences between mean and median values provide insights into the distributions of these attributes and the presence of potential outliers or asymmetry in the data.
```python
# Calculating the probability of a male customer buying a KP781 treadmill
total_customers = len(df)
male_customers = len(df[df['Gender'] == 'Male'])
kp781_customers = len(df[df['Product'] == 'KP781'])
male_and_kp781_customers = len(df[(df['Gender'] == 'Male') & (df['Product'] == 'KP781')])
```

```python
# Conditional probability: P(KP781 | Male) = P(KP781 and Male) / P(Male)
probability_kp781_given_male = male_and_kp781_customers / male_customers

print(f"Probability of a male customer buying a KP781 treadmill: {probability_kp781_given_male:.2%}")
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from yellowbrick.cluster import KElbowVisualizer

# Selecting relevant features for segmentation
features = df[['Age', 'Income', 'Fitness']]

# Standardizing the data (mean = 0, std = 1)
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)


# Determining the optimal number of clusters (k) using the Yellowbrick elbow method
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2, 11))  # Define the range of k values to explore
visualizer.fit(features_scaled)
visualizer.show()

# Optimal k value is determined visually from the elbow method plot
optimal_k = visualizer.elbow_value_
print(f'Optimal k value: {optimal_k}')

# Creating a K-means model with the optimal k value
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', max_iter=300, n_init=10, random_state=0)
cluster_labels = kmeans.fit_predict(features_scaled)

# Adding the cluster labels to the original dataset
df['Cluster'] = cluster_labels

# Visualizing the clusters using pair plots for the selected features
sns.pairplot(data=df, hue='Cluster', palette='Dark2')
plt.suptitle(f'Customer Segmentation (k={optimal_k})', y=1.02)
plt.show()
```

```python
# Calculating the total number of customers
total_customers = len(df)

# Calculating marginal probabilities for specific events
def calculate_marginal_probability(event):
    event_count = len(df[df['Product'] == event])
    marginal_probability = event_count / total_customers
    return marginal_probability

# Calculating conditional probability for specific events
def calculate_conditional_probability(event, condition):
    event_count = len(df[(df['Product'] == event) & (df['Gender'] == condition)])
    condition_count = len(df[df['Gender'] == condition])
    conditional_probability = event_count / condition_count
    return conditional_probability
```

```python
# Example 1: Marginal Probability of purchasing KP781
marginal_kp781_probability = calculate_marginal_probability('KP781')
print(f"Marginal Probability of purchasing KP781: {marginal_kp781_probability:.2%}")
```

```python
# Example 2: Conditional Probability of purchasing KP281 given Gender is Male
conditional_kp281_probability_male = calculate_conditional_probability('KP281', 'Male')
print(f"Conditional Probability of purchasing KP281 given Gender is Male: {conditional_kp281_probability_male:.2%}")
```
